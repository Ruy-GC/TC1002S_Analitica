```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("iris.csv")
dfset = df[df.Tipo_Flor == "Iris-setosa"]
dfvir = df[df.Tipo_Flor == "Iris-virginica"]
dfver = df[df.Tipo_Flor == "Iris-versicolor"]
```

```python
plt.scatter(dfset.Largo_sep,dfset.Ancho_sep)
plt.scatter(dfvir.Largo_sep,dfvir.Ancho_sep)
plt.scatter(dfver.Largo_sep,dfver.Ancho_sep)
plt.xlabel("largo_sep")
plt.ylabel("Ancho_sep")
plt.legend(["Iris-setosa","Iris-virginica","Iris-verdicolor"])
```

```python
plt.scatter(dfset.Largo_pet,dfset.Ancho_pet)
plt.scatter(dfvir.Largo_pet,dfvir.Ancho_pet)
plt.scatter(dfver.Largo_pet,dfver.Ancho_pet)
plt.xlabel("largo_pet")
plt.ylabel("Ancho_pet")
plt.legend(["Iris-setosa","Iris-virginica","Iris-verdicolor"])
```

```python
data = df.values
X = data[:,0:4]
y = data[:,[4]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

for k in range (3,12,2):
    neigh = KNeighborsClassifier(n_neighbors = k)
    neigh.fit(X_train,y_train)
    score = neigh.score(X_test,y_test)
    print(k,score)
    
```

```python
neigh.predict([[3,4.5,1,0.5]]) #setosa
```

```python
neigh.predict([[8,4,6,2]]) #virginica
```

```python
neigh.predict([[6,3,4,1.5]]) #versicolor
```
